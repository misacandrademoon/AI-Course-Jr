## 2. Identificación y Manejo de Valores Faltantes


### Identificar valores faltantes

```python
missing_values = df1.isnull().sum()

print("Valores faltantes por columna:")
print(missing_values)
```

¿Qué valores nulos encontraron y por qué decidieron imputarlos, eliminarlos o manejarlos de otra manera?
Los Valores nulos encontrados son Edad con 177 datos nulos, deck  con 688
y embark_town con solamente 2 valores nulos , se eliminara  la columna con mayordatos nulos que seria deck y de ahi usar funciones para rellenar los valores nulos respectivamente.
¿Qué cambios realizaron en los datos y cómo afectará esto su análisis posterior?


Decidimos:
- Imputar los valores de `age` utilizando la mediana para mantener la robustez ante outliers.
- Eliminar la columna `deck` debido a la gran cantidad de valores nulos.
- Imputar `embark_town` con el modo, ya que solo hay 2 valores nulos.


```python
# Imputar valores nulos
df1['age'].fillna(df1['age'].median(), inplace=True)
df1['embark_town'].fillna(df1['embark_town'].mode()[0], inplace=True)
df1.drop(columns='deck', inplace=True)
```

### Imputación de Valores Nulos
Realizamos la imputación de valores nulos para `age` y `embark_town`. La columna `deck` se eliminó debido a la gran cantidad de datos faltantes.



```python
missing_values = df1.isnull().sum()

print("Valores faltantes por columna:")
print(missing_values)
```

## 3. Transformación de datos:

#**Normalización:**

Escalar las variables numéricas (como tarifa y edad) usando métodos de normalización o estandarización.

```python
from sklearn.preprocessing import MinMaxScaler

# Normalizar las columnas 'age' y 'fare'
scaler = MinMaxScaler()
df1[['age', 'fare']] = scaler.fit_transform(df1[['age', 'fare']])
```

### Normalización
Normalizamos las variables `age` y `fare` para escalar sus valores entre 0 y 1. Esto ayuda a que las variables numéricas estén en un rango comparable, facilitando su uso en modelos de aprendizaje automático.


```python
# Codificación one-hot
df1 = pd.get_dummies(df1, columns=['sex', 'embarked', 'class', 'who', 'embark_town'])
```

### Codificación One-Hot
Aplicamos codificación one-hot a las variables categóricas para transformarlas en un formato adecuado para el análisis y modelado posterior. Esto es esencial para que los modelos puedan procesar estas variables de manera efectiva.



#Análisis Exploratorio de Datos (EDA):

```python
# Descripción de las variables
df1.describe()

# Graficar distribuciones
sns.histplot(df1['age'], kde=True)
plt.title('Distribución de Edad')
plt.show()

sns.histplot(df1['fare'], kde=True)
plt.title('Distribución de Tarifa')
plt.show()

# Heatmap de correlaciones
# Select only numeric columns for correlation calculation
numeric_df1 = df1.select_dtypes(include=['number'])
sns.heatmap(numeric_df1.corr(), annot=True, cmap='coolwarm')
plt.title('Mapa de Correlaciones')
plt.show()

```

### Análisis Exploratorio de Datos (EDA)
Observamos las distribuciones de las variables principales como `age` y `fare`, y creamos un mapa de calor para visualizar las correlaciones entre las variables.
